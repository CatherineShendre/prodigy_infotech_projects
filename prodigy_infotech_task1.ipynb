#  **DROP** **UNUSEFUL** **DATA** ✅

---





```python
data = data.drop('Unnamed: 0', axis=1)
```

#  **MOST POPULATED COUNTRIES IN YEAR 1960 AND 2020** ✅



```python
#TOP TEN MOST POPULATED COUNTIRES IN YEAR 1960
top_ten = data.sort_values('year_1960', ascending=False).head(10)
plt.figure(figsize=(15, 6))
plt.bar(top_ten['country'], top_ten['year_1960'])
plt.xlabel('Country')
plt.ylabel('Population (In Billions)')
plt.title('Top 10 Most Populated Countries in 1960')
plt.show()

#TOP TEN MOST POPULATED COUNTIRES IN YEAR 2020
top_ten = data.sort_values('year_2020', ascending=False).head(10)
plt.figure(figsize=(15, 6))
plt.bar(top_ten['country'], top_ten['year_2020'])
plt.xlabel('Country')
plt.ylabel('Population (In Billions)')
plt.title('Top 10 Most Populated Countries in 2020')
plt.show()
```

#  **LEAST POPULATED COUNTRIES IN YEAR 1960 AND 2020** ✅

```python
#LEAST TEN MOST POPULATED COUNTIRES IN YEAR 1960
bottom_ten = data.sort_values('year_1960', ascending=True).head(10)
plt.figure(figsize=(15, 6))
plt.bar(bottom_ten['country'], bottom_ten['year_1960'])
plt.xlabel('Country')
plt.ylabel('Population (In Billions)')
plt.title('bottom 10 Most Populated Countries in 1960')
plt.show()

#LEAST TEN MOST POPULATED COUNTIRES IN YEAR 1960
bottom_ten = data.sort_values('year_2020', ascending=True).head(10)
plt.figure(figsize=(15, 6))
plt.bar(bottom_ten['country'], bottom_ten['year_2020'])
plt.xlabel('Country')
plt.ylabel('Population (In Billions)')
plt.title('bottom 10 Most Populated Countries in 2020')
plt.show()



```

#  **CHINA'S POPULATION GROWTH** ✅


---



```python
country = 'China'
plt.plot(data.loc[data['country'] == country].iloc[:, 1:].values.T, label=country)
plt.xlabel('Year')
plt.ylabel('Population (In Billions)')
plt.title('Population Growth over Time: {country}')
plt.legend()
plt.show()
```

#  **INDIA'S POPULATION GROWTH** ✅

---



```python
country = 'India'
plt.plot(data.loc[data['country'] == country].iloc[:, 1:].values.T, label=country)
plt.xlabel('Year')
plt.ylabel('Population (In Billions)')
plt.title('Population Growth over Time: {country}')
plt.legend()
plt.show()
```

# **POPULATION COMPARISON BETWEEN MOST POPULATED COUNTRIES INDIA AND CHINA** ✅

---



```python
countries = ['India', 'China']
populations = [data.loc[data['country'] == 'India', 'year_1960'].iloc[0],
               data.loc[data['country'] == 'China', 'year_1960'].iloc[0]]
plt.figure(figsize=(8, 4))
plt.bar(countries, populations)
plt.title('Population comparison between India and China in 1960')
plt.xlabel('Country')
plt.ylabel('Population (in millions)')
plt.show()


countries = ['India', 'China']
populations = [data.loc[data['country'] == 'India', 'year_2020'].iloc[0],
               data.loc[data['country'] == 'China', 'year_2020'].iloc[0]]
plt.figure(figsize=(8, 4))
plt.bar(countries, populations)
plt.title('Population comparison between India and China in 2020')
plt.xlabel('Country')
plt.ylabel('Population (in millions)')
plt.show()

```
